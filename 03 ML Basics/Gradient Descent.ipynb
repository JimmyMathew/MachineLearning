# Notebook imports and packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example1 - A simple cost function

## $f(x) = x^2 + x +1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#Make data 
x_1 = np.linspace(start=-3,stop=3,num=500) # Creates a sequence of numbers using numpy
```

```python
#Plot
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Solpes & Derivatives

```python
#derivative of f(x)
def df(x):
    return 2*x + 1
```

```python
#Plot function and derivative side by side
plt.figure(figsize=[15,5])#measured in inches (width and height)

#chart:1 - Cost function 
plt.subplot(1,2,1)#row,column,index 
#plt.subplot(2,1,1)#To stack it one after another instead of side by side 
plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('Cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1, f(x_1), color='blue',linewidth=3)

#chart:2 - Derivative function
plt.subplot(1,2,2)#row,column,index 
#plt.subplot(2,1,2)#To stack it one after another instead of side by side
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.title('Slope of the Cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()
plt.plot(x_1,df(x_1),color='skyblue',linewidth=5)
plt.show()
```

   ## Gradient Descent

```python
new_x = 3 #Starting point
previous_x = 0 #temp variable used inside the loop
step_multiplier = 0.1 #Learning rate
precision = 0.00001

x_list = [new_x]
slope_list  = [df(new_x)]
for n in range(500):
    previous_x = new_x #Make a guess or prediction to start with
    gradient = df(previous_x)#calculate the error - The steepness of slope tells us how far off we are from the minimum. Because at the minimum the slope = 0 
    new_x = previous_x - step_multiplier * gradient  #Learning part by moving downward
    step_size = abs(new_x - previous_x)
    #print('step_size: ', step_size)
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if step_size < precision:
        print('Loop ran this many times:',n)
        break
print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is: ', df(new_x))
print('f(x) or cost at this point is: ',f(new_x))
    
```

```python
#Superimpose the gradient descent calculations
plt.figure(figsize=[20,5])#measured in inches (width and height)

#chart:1 - Cost function 
plt.subplot(1,3,1)#row,column,index 
#plt.subplot(2,1,1)#To stack it one after another instead of side by side 
plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('Cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1, f(x_1), color='blue',linewidth=3,alpha=0.8)
values = np.array(x_list)
plt.scatter(x_list,f(values),color='red',s=100,alpha=0.6)#To point the values along the plot

#chart:2 - Derivative function
plt.subplot(1,3,2)#row,column,index 
#plt.subplot(2,1,2)#To stack it one after another instead of side by side
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.title('Slope of the Cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()
plt.plot(x_1,df(x_1),color='skyblue',linewidth=5,alpha=0.6)
plt.scatter(x_list,slope_list,color='red',s=100,alpha=0.6)#To point the values along the plot


#chart:3 - Derivative (Close Up)
plt.subplot(1,3,3)#row,column,index 
#plt.subplot(2,1,2)#To stack it one after another instead of side by side
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.title('Gradient Descent (close up)',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.grid()
plt.plot(x_1,df(x_1),color='skyblue',linewidth=6,alpha=0.8)
plt.scatter(x_list,slope_list,color='red',s=300,alpha=0.6)#To point the values along the plot
plt.show()
```
