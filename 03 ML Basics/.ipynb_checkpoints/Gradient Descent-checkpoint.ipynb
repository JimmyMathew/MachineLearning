# Notebook imports and packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example1 - A simple cost function

## $f(x) = x^2 + x +1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#Make data 
x_1 = np.linspace(start=-3,stop=3,num=500) # Creates a sequence of numbers using numpy
```

```python
#Plot
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Solpes & Derivatives

```python
#derivative of f(x)
def df(x):
    return 2*x + 1
```

```python
#Plot function and derivative side by side
plt.figure(figsize=[15,5])#measured in inches (width and height)

plt.subplot(1,2,1)#row,column,index 
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1, f(x_1))
plt.subplot(1,2,2)#row,column,index 
plt.plot(x_1,df(x_1))
plt.show()
```
