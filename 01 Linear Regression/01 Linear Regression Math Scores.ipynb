```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('lsd_math_score_data.csv')
```

```python
time =  data[['Time_Delay_in_Minutes']]
lsd = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
#To export the chart separately along with the jupyter notbook when we download we use the below command(jupyter specific)
%matplotlib inline 

plt.title('Tissue concentration of LSD over time',fontsize=17)
plt.xlabel('Time in minutes',fontsize=14)
plt.ylabel('Tissue LSD ppm',fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.text(x=0,y=-0.5,s='Wagner et al. (1968)')

plt.ylim(1, 7)
plt.xlim(0, 500)
plt.style.use('classic')#custom chart styles 
plt.plot(time,lsd,color='#e74c3c',linewidth=3)#x axis is time and y axis is lsd
plt.show()
```

```python
regression = LinearRegression()
regression.fit(lsd,score)#Run Linear Regression 
print('Theta1:',regression.coef_[0][0])
print('Theta0 (or) Intercept:', regression.intercept_[0])
print('R-Square :' ,regression.score(lsd,score))
predicted_score = regression.predict(lsd)
```

Visualizing the line drawn by LR

```python
%matplotlib inline 
plt.title('Arithemetic vs LSD-25',fontsize=17)
plt.xlabel('Tissue LSD ppm',fontsize=14)
plt.ylabel('Performance Score',fontsize=14)
plt.ylim(25, 85)
plt.xlim(1, 6.5)
plt.style.use('fivethirtyeight')
plt.scatter(lsd,score,color='blue',s=100,alpha=0.7)
plt.plot(lsd,predicted_score,color='red',linewidth=3)
plt.show()
```
